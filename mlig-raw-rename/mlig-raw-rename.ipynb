## Naming convention

The naming convention will be: species, experiment, treatment, time, sample, replicate. All fields separate with an underscore (_). 

For example: 

    mlig_regen_reg_072h_1_1
    mlig_regen_reg_072h_2_1
    mlig_regen_reg_072h_3_1
    mlig_regen_reg_072h_3_2
    
The above 4 libraries are twice amputated worms at 72 hours after initial cutting. There are 3 biological samples, where the first two (1, 2) have no replicates and the third has two replicates.

## Filter for relevant files

```python
meta = [(p.split('/')[1:3],p.split('/')[-1]) for p in paths]
filtered = [','.join(x[0]) + ' - ' + x[1] for x in meta if not x[0][0][0].isdigit()]
cleaned = [','.join(x[0]) + ' - ' + x[1] for x in meta if x[0][0][0].isdigit()]

print 'Clean up path data to get relevant information:\n'
print 'filtered out:(' + str(len(filtered)) + ')'
print '\n'.join(filtered)
print
print 'kept:(' + str(len(cleaned)) + ')'
print '\n'.join(cleaned)
```

```python
meta = [p.split('/') for p in paths if p.split('/')[1][0].isdigit()]
cleaned = []
print 'Filtered out 0 hour pair-end samples:(4)'
for item in meta:
    if item[1] == '0h_control':
         if item[2].find('tech') == -1:
                print item[1],item[2],item[-3],item[-1]
                continue
    cleaned.append(item)
print
print "Final number of fastq.gz files:", len(cleaned)
```

## Parse path and file names to generate new names

```python
#parse time-point, condition, replicates
#output for spotcheck
data = {}
dayMap = {0:tps[0],12:tps[1],24:tps[2],2:tps[3],3:tps[4],4:tps[5],5:tps[6],7:tps[7]}
counts = {}
index = {}
for p in cleaned:
    day = p[1].split('_')[0]
    if not day[-1].isdigit():
        day = day[:-1]
    day = dayMap[int(day)]
    con = -1
    if p[1].split('_')[-1] == 'control':
        con = 'control'
    elif p[1].split('_')[-1] == 'reg':
        con = 'reg'
        
    if p[2].split('_')[-1] == 'control':
        con = 'control'
    elif p[2].split('_')[-1] == 'reg':
        con = 'reg'
    
    tech = False
    if p[-3].find('tech') != -1:
        tech = True
    
    prefix = 'mlig_regen_' + con + '_' + day + '_'
    if not counts.has_key(prefix):
        counts[prefix] = [False,False]
        
    if tech:
        if not counts[prefix][1]:
            print '/'.join(p)
            print prefix + '3_1'
            index['/'.join(p)] = prefix + '3_1'
            counts[prefix][1] = True
        else:
            print '/'.join(p)
            print prefix + '3_2'
            index['/'.join(p)] = prefix + '3_2'
    else:
        if not counts[prefix][0]:
            print '/'.join(p)
            print prefix + '1_1'
            index['/'.join(p)] = prefix + '1_1'
            counts[prefix][0] = True
        else:
            print '/'.join(p)
            print prefix + '2_1'
            index['/'.join(p)] = prefix + '2_1'
    print
```

### Check for duplicate new names

4 libraries contained duplicated new names. These files contain 3 bio replicates only instead of 3 bio replicates and an extra technical replicate. 

```python
print 'Number of files and new names: ', len(index.items())
from collections import Counter
count = Counter([x[1] for x in index.items()])
dup = [x[0] for x in count.items() if x[1] > 1]
print 'Files with duplicate new names:\n' + '\n'.join(dup)
print
dupPaths = [x for x in index.items() if x[1] in dup]
dupPaths.sort(key = lambda x : x[1])
rename = [x[0] for x in dupPaths[::2]]
print 'Changing the following paths to a new name:\n'
for name in rename:
    old = index[name]
    meta = index[name].split('_')
    meta[-2] = '3'
    newName = '_'.join(meta)
    index[name] = newName
    print name
    print newName
    print 

count = Counter([x[1] for x in index.items()])
dup = [x[0] for x in count.items() if x[1] > 1]
print 'Rechecking for duplicates:', len(dup)
print 'Final number of files with new names: ', len(index.items())
```
