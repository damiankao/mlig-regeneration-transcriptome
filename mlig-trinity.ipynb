 3. Unstranded assembly
 
        Trinity --seqType fq --JM 50G --left /mlig_working/reads/left.stranded.fastq --right /mlig_working/reads/right.stranded.fastq --CPU 32

### Stranded assembly vs unstranded assembly

Map stranded libraries to both stranded and unstranded assembly. Check to see how many transcripts had reads of only one orientation mapping vs reads of both orientations mapping.

```python
fs = [x for x in os.listdir('/home/share/ngsData/mlig/') if x.split('.')[-1] == 'gz' and x.find('pilot') != -1]
template = 'bowtie2 --no-unal --no-hd --very-fast -X 1500 -p 32 -x BTINDEX -1 FASTQ1 -2 FASTQ2 > '
cmds = ''
for lib in list(set(['_'.join(f.split('_')[:-1]) for f in fs if f.split('.')[0].split('_')[-1][0] == 'R'])):
    if lib.find('_A') == -1:
        a = lib + '_R1.fastq.gz'
        b = lib + '_R2.fastq.gz'
        a = a + '.paired.trimmed.fastq.gz'
        b = b + '.paired.trimmed.fastq.gz'
        cmds += template.replace('BTINDEX','mlig.stranded').replace('FASTQ1',a).replace('FASTQ2',b) + lib + '.stranded.sam\n'
        cmds += template.replace('BTINDEX','mlig.unstranded').replace('FASTQ1',a).replace('FASTQ2',b) + lib + '.unstranded.sam\n'
HTML(textarea('Bowtie2 mapping commands',cmds))
```

```python
#for each transcript, record number of PE reads in either directions

HTML(textarea('Script for recording PE read direction from .SAM file','''import sys, os
from collections import defaultdict

fs = [x for x in os.listdir('.') if x.split('.')[-1] == 'sam']

for f in fs:
    tids = defaultdict(lambda : [0,0])
    inFile = open(f)
    outFile = open('.'.join(f.split('.')[:-1]) + '.strand','w')
    for line in inFile:
        data = line.strip().split('\\t')
        if data[1] == '99':
            tids[data[2]][0] += 1
        elif data[1] == '163':
            tids[data[2]][1] += 1
            
    for tid, count in tids.items():
        outFile.write(tid + '\\t' + str(count[0]) + '\\t' + str(count[1]) + '\\n')
    inFile.close()
    outFile.close()'''))
```

```python
prefix = '/home/share/projects/mlig_transcriptome/strandness/'
strandFs = [x for x in os.listdir(prefix) if x.split('.')[-1] == 'strand']
from collections import defaultdict
strandRes = defaultdict(lambda : [0,0])
for strandF in strandFs:
    inFile = open(prefix + strandF)
    
    for line in inFile:
        meta = line.strip().split('\t')
        a = int(meta[1])
        b = int(meta[2])
        if min(a,b) == 0 and max(a,b) > 5:
            strandRes[strandF][0] += 1
        elif a > 5 and b > 5:
            strandRes[strandF][1] += 1
    
    inFile.close()
    
strandRes = strandRes.items()
strandRes.sort(key = lambda x : x[0])
```

```python
strandTable = ListTable()
strandTable.append(['Library','Single','Mixed','Single %','Mixed %'])
for item in strandRes:
    if item[0].split('.')[1] == 'stranded':
        singleP = item[1][0] / 226062.0 * 100
        mixedP = item[1][1] / 226062.0 * 100
        
        strandTable.append([item[0].split('.')[0] + ', ' + item[0].split('.')[1] + ' assembly',commas(item[1][0]),commas(item[1][1]),singleP,mixedP])
    else:
        singleP = item[1][0] / 189102.0 * 100
        mixedP = item[1][1] / 189102.0 * 100
        
        strandTable.append([item[0].split('.')[0] + ', ' + item[0].split('.')[1] + ' assembly',commas(item[1][0]),commas(item[1][1]),singleP,mixedP])
        
        
print 'Number of transcripts with single orientation versus mixed orientation for each library:'
print
print '*Single means the transcript has more than 5 PE reads in one direction and 0 PE reads in the other direction.'
print '*Mixed means the transcript has more than 5 PE reads in either directions'
print
print '*Total number of transcripts for stranded assembly: 226,062'
print '*Total number of transcripts for unstranded assembly: 189,102'
strandTable
```

### Effects of Jaccard clipping 

Generated two assemblies with the same input (stranded, PE reads). One assembly without jaccard clipping option turned on and one with jaccard clipping option turned on.

1. Check transcriptome metrics (length, N50..)

2. Blast the two assemblies against complete proteomes. Check for potential fusion events.

    blastx -query combined.fa -db proteomes -evalue 1e-5 -outfmt '6 qseqid sseqid qlen slen pident length qstart qend sstart send evalue' -num_threads 10 > jaccard.blast.out
    
    Use the following scripts to parse the blast output file to generate:
 
     - A bed file where the query and subject ids are merged to get union of all HSPs for a hit
     - A bed file for just query ids to get union of all hits

3. Use TransDecoder and see the distribution of number of predicted peptides for each transcript.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

noJaccardFa = open('/home/share/projects/mlig_transcriptome/checkJaccard/mlig.stranded.fasta')
jaccardFa = open('/home/share/projects/mlig_transcriptome/checkJaccard/mlig.strandedJaccard.fasta')

noJaccardLengths = []
jaccardLengths = []

for line in noJaccardFa:
    if line[0] == '>':
        l = int(line.split()[1].split('=')[1])
        noJaccardLengths.append(l)
        
for line in jaccardFa:
    if line[0] == '>':
        l = int(line.split()[1].split('=')[1])
        jaccardLengths.append(l)

jaccardLengths = pd.Series(jaccardLengths)
noJaccardLengths = pd.Series(noJaccardLengths)
```

```python
print 'no jaccard clipping assembly:'
print ' mean length:', noJaccardLengths.mean()
print ' median length:', noJaccardLengths.median()
print 'jaccard clipping assembly:'
print ' mean length:', jaccardLengths.mean()
print ' median length:', jaccardLengths.median()

fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = [15,5])
noJaccardLengths[noJaccardLengths < 2000].hist(bins=100, ax = axes[0])
axes[0].set_title('no jaccard clipping assembly')
jaccardLengths[jaccardLengths < 2000].hist(bins=100, ax = axes[1])
axes[1].set_title('jaccard clipping assembly')
fig.tight_layout()
```

```python
#format the blast file into a sorted .bed file for bedtools merge
HTML(textarea('Create query-subject .bed file for HSP unions','''import sys
from collections import defaultdict

jaccardBlast = open(sys.argv[1])

transcript = defaultdict(list)
for line in jaccardBlast:
    data = line.strip().split()
    transcript[data[0] + '.' + data[1]].append((data[-3],data[-2],data[1],data[-1]))
    
tids = transcript.keys()
#sort by "chromosome"
tids.sort()

for tid in tids:
    feats = transcript[tid]
    #sort by start coodinate
    feats.sort(key = lambda x : int(x[0]))
    for feat in feats:
        print tid + '\\t' + '\\t'.join(feat)'''))
```

```python
#format the blast file into a sorted .bed file for bedtools merge
HTML(textarea('Create query .bed file for hit unions','''import sys
from collections import defaultdict

jaccardBlast = open(sys.argv[1])

transcript = defaultdict(list)
for line in jaccardBlast:
    data = line.strip().split()
    transcript[data[0]].append((data[-3],data[-2],data[1],data[-1]))
    
tids = transcript.keys()
#sort by "chromosome"
tids.sort()

for tid in tids:
    feats = transcript[tid]
    #sort by start coodinate
    feats.sort(key = lambda x : int(x[0]))
    for feat in feats:
        print tid + '\\t' + '\\t'.join(feat)'''))
```

```python
hitFile = open('/home/share/projects/mlig_transcriptome/checkJaccard/jaccard.blast.hit.merged.out')
hspFile = open('/home/share/projects/mlig_transcriptome/checkJaccard/jaccard.blast.hsp.merged.out')

unions = defaultdict(int)
lengths = defaultdict(lambda : defaultdict(int))

for line in hitFile:
    meta = line.strip().split()
    unions[meta[0]] += int(meta[2]) - int(meta[1]) + 1
    
for line in hspFile:
    meta = line.strip().split()
    q = '.'.join(meta[0].split('.')[:2])
    s = meta[0].split('.')[2]
    lengths[q][s] += int(meta[2]) - int(meta[1]) + 1    

hitFile.close()
hspFile.close()

jaccardScore = []
noJaccardScore = []
jaccardIndex = []
noJaccardIndex = []
for q, ls in lengths.items():
    fscore = max(ls.values()) / float(unions[q])
    if q[0] == 'j':
        jaccardScore.append(fscore)
        jaccardIndex.append(q)
    else:
        noJaccardScore.append(fscore)
        noJaccardIndex.append(q)
```

```python
jaccardScore = pd.Series(jaccardScore, index = jaccardIndex)
noJaccardScore = pd.Series(noJaccardScore, index = noJaccardIndex)
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = [15,5])
noJaccardScore.hist(bins = 100, ax = axes[0])
jaccardScore.hist(bins = 100, ax = axes[1])

axes[0].set_title('no jaccard clipping fusion score distribution')
axes[0].set_title('jaccard clipping fusion score distribution')

fig.tight_layout()
```

```python
transPep = open('/home/share/projects/mlig_transcriptome/checkJaccard/transdecoder/combined.fasta.transdecoder.pep')

jaccardTrans = defaultdict(int)
noJaccardTrans = defaultdict(int)

for line in transPep:
    if line[0] == '>':
        tid = line[1:].strip().split('|')[0]
        if tid[0] == 'j':
            jaccardTrans[tid] += 1
        else:
            noJaccardTrans[tid] += 1
            
jaccardTrans = pd.Series(jaccardTrans)
noJaccardTrans = pd.Series(noJaccardTrans)
```
