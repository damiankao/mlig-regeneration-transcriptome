# Read saturation analysis

The purpose of this analysis is to see if there is a saturation in blast hits of assembled contigs (with Inchworm) as we increase reads. The following was performed (consolidating and shuffling was done in the data processing notebook):

 1. Consolidation of all reads into one file. Since inchworm (greedy extension) does not perform scaffolding, PE reads were designated as two single end reads.
 2. This file is shuffled.
 3. The reads are binned into 50 million reads bins.
 4. For increaseing 50 million reads, an inchworm assembly is performed

I only used contigs longer than 300 bases for computational reasons. A single inchworm assembly with all availble reads yielded 19,118,230 contigs, which will just take too long to blast.

By filtering for 300 bases, the combined total of all contigs from every 50 million bin was only 2,001,998 contigs. A much easier amount to blast against the complete proteomes. 

```python
HTML(textarea('Script for filtering inchworm contigs for length','''
import sys,os
from Bio import SeqIO

fs = [x for x in os.listdir('.') if x.split('.')[-1] == 'fa']

for f in fs:
    count = 0
    inFile = open(f)
    name = 'bin.' + f.split('.')[-2]

    for record in SeqIO.parse(inFile,'fasta'):
        if len(str(record.seq)) >= 300:
            print '>' + name + '.' + str(count)
            print str(record.seq)
            count += 1
    inFile.close()
'''))
```

```python
from collections import defaultdict
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

prefix = '/home/share/projects/mlig_transcriptome/readSaturation/blast/'
eRange = [5,10,15,20,25,50,100]
orgs = {}
for f in os.listdir(prefix):
    inFile = open(prefix + f)
    
    blastBin = {}
    for i in eRange:
        blastBin[i] = defaultdict(list)
    
    for line in inFile:
        meta = line.strip().split()
        bid = meta[1].split('.')[1]
        evalue = float(meta[-1])
        tid = meta[0]
        
        for i in eRange:
            if evalue < float("1e-" + str(i)):
                blastBin[i][bid].append(tid)
    
    orgs[f.split('.')[0]] = blastBin
    inFile.close()
```

### Ortholog hit ratio saturation

For each protein/transcript in the complete proteomes/transcriptomes:

 1. Filter for proteins that has a hit in all 22 bins
 
 2. Find the largest ortholog hit ratio (alignment length / protein length) for each bin
 
 3. Plot the largest ratio for each species bins

```python
from collections import defaultdict
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

prefix = '/home/share/projects/mlig_transcriptome/readSaturation/blast/'
fs = [x for x in os.listdir(prefix) if x.split('.')[-1] == 'out']
orgs = {}
for f in fs:
    inFile = open(prefix + f)
    sids = defaultdict(lambda : defaultdict(float))
    for line in inFile:
        meta = line.strip().split()
        binNum = meta[1].split('.')[1]
        ratio = (float(meta[7]) - float(meta[6]) + 1) / float(meta[2])
        if sids[meta[0]][binNum] < ratio:
            sids[meta[0]][binNum] = ratio
            
    ratioSums = defaultdict(list)
    for sid, data in sids.items():
        if len(data) == 22:
            for i in range(22):
                ratioSums[i].append(data[str(i)])

    ratioSums = ratioSums.items()
    ratioSums.sort(key = lambda x : x[0])
    orgs[f.split('.')[0]] = [sum(x[1]) / len(x[1]) for x in ratioSums]
```

```python
import math
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=[20,10])

print 'BLAST ortholog hit ratio with increasing reads (50 million) against various species'
speNames = {'smed':'S. mediterranea','csin':'C. sinensis','emul':'E. multilocularis','hsap':'H. sapiens','cele':'C. elegans','sman':'S. mansoni','dmel':'D. melanogaster',}
for i, spe in enumerate(speNames.keys()):
    pd.Series(orgs[spe]).plot(ax = axes[math.ceil(i/3),i % 3])
    axes[math.ceil(i/3),i % 3].set_title(speNames[spe])
    axes[math.ceil(i/3),i % 3].set_xlabel('50 million reads bins')
    axes[math.ceil(i/3),i % 3].set_ylabel('Number of hits (eval < 1e-5)')
fig.tight_layout()
```
