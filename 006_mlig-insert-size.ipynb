# Check insert sizes of pair-end reads

All pair-end reads were mapped to an inchworm assembly performed on all reads. Only pair-end reads where both ends mapped uniquely and correctly were kept to look at insert sizes.

```python
import gzip, os, sys
#import custom functions for displaying tables, bash commands
sys.path.append(os.path.abspath("/home/damian/"))
from dk_ipython import *
from IPython.display import HTML
import json
```

```python
template = 'bowtie2 --very-fast -X 1500 --no-mixed --no-discordant --no-dovetail --no-contain --no-overlap -p 32  -x /mlig_insert/index/mlig.inchworm.K25.L25.DS.21 -1 FASTQ1 -2 FASTQ2 | samtools view -bS -F 4 - > '
cmds = ''
trimmLog = open('resource/trimPE.err')
fs = []
for line in trimmLog:
    if line.find('Started with arguments') != -1:
        prev = line.strip().split('.fastq.gz')[0].split()[-1].split('/')[-1][:-1]
        fs.append(prev)
        cmds += template.replace('FASTQ1','/mlig_raw/trim/' + prev + '1.fastq.gz.paired.trimmed.fastq.gz').replace('FASTQ2','/mlig_raw/trim/' + prev + '2.fastq.gz.paired.trimmed.fastq.gz') + prev + '.bam\n'
trimmLog.close()
HTML(textarea('bowtie2 commands',cmds))
```

```python
HTML(textarea('Script for extracing insert lengths from .SAM file','''
import sys,os

fs = [x for x in os.listdir('.') if x.split('.')[-1] == 'sam']

for f in fs:
    outF = open(f.split('.')[0] + '.insert','w')
    for line in open(f,'r'):
        if line[0] != '@':
            meta = line.strip().split()
            if meta[2] != '*':
                outF.write(str(abs(int(meta[8]))) + '\\n')
    outF.close()
'''))
```

### Insert size distributions

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

prefix = '/home/share/projects/mlig_transcriptome/fragmentSize/size/'
insertFs = os.listdir(prefix)

inserts = {}
for f in insertFs:
    lengths = pd.Series([int(x) for x in open(prefix + f)])
    inserts[f.split('.')[0]] = lengths
```

```python
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=[20,20])
keys = inserts.keys()
keys.sort()
index = 0
for row in range(4):
    for col in range(4):
        if index < len(keys):
            inserts[keys[index]].hist(bins = 100, ax = axes[row, col], alpha=0.8)
            axes[row,col].set_title(keys[index])
            index += 1
```
