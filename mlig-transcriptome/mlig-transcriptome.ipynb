### Trimming M. Lignano reads with trimmomatic

```python
import gzip, os, sys
#import custom functions for displaying tables, bash commands
sys.path.append(os.path.abspath("/home/damian/"))
from dk_ipython import *
from IPython.display import HTML

#trim all mlig libraries for quality of 30 and min length of 40
fs = [x for x in os.listdir('/home/share/ngsData/mlig/') if x.split('.')[-1] == 'gz']

#done on AWS. Adapters were generated from concatenating given adapters in Trimmomatic.
#generate trimmmomatic commands and run with:
#parallel :::: trimm.cmd > trimm.out 2> trimm.err 

trimCmdSEA = 'java -jar /mlig_raw/software/Trimmomatic-0.32/trimmomatic-0.32.jar SE -threads 1 '
trimCmdPEA = 'java -jar /mlig_raw/software/Trimmomatic-0.32/trimmomatic-0.32.jar PE -threads 1 '
trimCmdB = ' LEADING:30 TRAILING:30 MINLEN:40 ILLUMINACLIP:/mlig_raw/software/Trimmomatic-0.32/adapters/adapters.fa:2:30:12:1:false TOPHRED33'
trimCmds = ''
for lib in list(set(['_'.join(f.split('_')[:-1]) for f in fs if f.split('.')[0].split('_')[-1][0] == 'R'])):
    a = lib + '_R1.fastq.gz'
    b = lib + '_R2.fastq.gz'
    phred = detectPhred('/home/share/ngsData/mlig/' + a, 5000)
    
    if phred == 'Sanger':
        phred = '-phred33 '
    else:
        phred = '-phred64 '
        
    a = '/mlig_raw/' + lib + '_R1.fastq.gz'
    b = '/mlig_raw/' + lib + '_R2.fastq.gz'
    #paired end
    trimCmds += trimCmdPEA + phred + a + ' ' + b + ' ' + \
    a + '.paired.trimmed.fastq.gz ' + a + '.unpaired.trimmed.fastq.gz ' + \
    b + '.paired.trimmed.fastq.gz ' + b + '.unpaired.trimmed.fastq.gz' + trimCmdB + '\n'
    
for lib in [f for f in fs if f.split('.')[0].split('_')[-1][0] != 'R']:
    lib = lib.split('.')[0]
    phred = detectPhred('/home/share/ngsData/mlig/' + lib + ".fastq.gz", 5000)
    if phred == 'Sanger':
        phred = '-phred33 '
    else:
        phred = '-phred64 '
        
    trimCmds += trimCmdSEA + phred + '/mlig_raw/' + lib + '.fastq.gz ' + lib + '.trimmed.fastq.gz' + trimCmdB + '\n'
    
HTML(textarea('Trimmomatic bash commands',trimCmds))
```

### Trimming results

```python
#load trimming log file (trimm.err) and check
trimmLib = {}
prev = ''
trimmLog = open('trim.err')
for line in trimmLog:
    if line.find('Started with arguments') != -1:
        prev = (line.strip().split('/EBS/')[1].split('.fastq.gz')[0], line.strip().split(':')[0][-2:])
    if line.find('Surviving') != -1:
        meta = line.strip().split(':')
        numReads = int(meta[1].split()[0])
        keys = ['Reads'] + [meta[1].split()[-1]] + [' '.join(x.split()[2:]) for x in meta[2:-1]]
        vals = [(numReads,100)] + [(int(x.split()[:2][0]),float(x.split()[:2][1][1:-2])) for x in meta[2:]]
        kv = dict(zip(keys,vals))
        kv['type'] = prev[1]
        trimmLib[prev[0]] = kv

trimmTable = ListTable()
trimmTable.append(['Library','# reads','Surviving','Dropped','Dropped Percent','Lib Type'])
trimmTable.extend([[k, commas(v['Reads'][0]),commas(v['Surviving'][0]),commas(v['Dropped'][0]),v['Dropped'][1],v['type']] for k,v in trimmLib.items()])
trimmTable
```

```python
import json
def genScript(trimInfo, prefix, libNames = []):
    pairA = []
    pairB = []
    se = []
    if len(libNames) == 0:
        libNames = trimInfo.keys()
    for libName, libStat in trimInfo.items():
        if libName in libNames:
            if libStat['Dropped'][1] != 100.0:
                if libStat['type'] == 'PE':
                    pairA.append(libName[:-1] + '1.fastq.gz.paired.trimmed.fastq.gz')
                    se.append(libName[:-1] + '1.fastq.gz.unpaired.trimmed.fastq.gz')
                    pairB.append(libName[:-1] + '2.fastq.gz.paired.trimmed.fastq.gz')
                    se.append(libName[:-1] + '2.fastq.gz.unpaired.trimmed.fastq.gz')
    for libName, libStat in trimInfo.items():
        if libName in libNames:
            if libStat['Dropped'][1] != 100.0:
                if libStat['type'] == 'SE':
                    se.append(libName + '.trimmed.fastq.gz')
        
    #generate script to make left.fastq, right.fastq, both.fa. left/right.fastq can be used for jaccard clipping and both.fa is used for trinity
    return '''
import gzip
pairA = ''' + json.dumps(pairA) + '''
pairB = ''' + json.dumps(pairB) + '''
se = ''' + json.dumps(se) + '''
outA = open("''' + prefix +  '''.left.fastq",'w')
outB = open("''' + prefix +  '''.right.fastq",'w')
outBoth = open("''' + prefix +  '''.both.fa",'w')
aCount = 0
for a in pairA:
    name = a.split('.')[0]
    if a.split('.')[0].split('_')[-1][0] == 'R':
        name = name[:-1]

    stream = gzip.open('/mlig_raw/trim/' + a,'rb')
    for line in stream:
        seq = stream.next()
        stream.next()
        qual = stream.next()
        outA.write('@r' + str(aCount).zfill(10) + '_' + name + '/1\\n' + seq + '+\\n' + qual)
        outBoth.write('>r' + str(aCount).zfill(10) + '_' + name + '/1\\n' + seq)
        aCount += 1
        
for a in se:
    name = a.split('.')[0]
    if a.split('.')[0].split('_')[-1][0] == 'R':
        name = name[:-1]

    stream = gzip.open('/mlig_raw/trim/' + a,'rb')
    for line in stream:
        seq = stream.next()
        stream.next()
        qual = stream.next()
        outBoth.write('>r' + str(aCount).zfill(10) + '_' + name + '/1\\n' + seq)
        aCount += 1

bCount = 0
for b in pairB:
    name = b.split('.')[0]
    if b.split('.')[0].split('_')[-1][0] == 'R':
        name = name[:-1]

    stream = gzip.open('/mlig_raw/trim/' + b,'rb')
    for line in stream:
        seq = stream.next()
        stream.next()
        qual = stream.next()
        outB.write('@r' + str(bCount).zfill(10) + '_' + name + '/2\\n' + seq + '+\\n' + qual)
        outBoth.write('>r' + str(bCount).zfill(10) + '_' + name + '/2\\n' + seq)
        bCount += 1

outA.close()
outB.close()
outBoth.close()
'''

HTML(textarea('Python script for making fasta file for Trinity',genScript(trimmLib, 'mlig')))
```

Run the following commands to first setup required software:

    sudo apt-get install -y make g++ zlib1g-dev default-jre default-jdk unzip ncurses-dev

    curl -o trinity.tar.gz -L http://sourceforge.net/projects/trinityrnaseq/files/latest/download?source=files
    curl -o bowtie.zip -L http://sourceforge.net/projects/bowtie-bio/files/bowtie/1.0.1/bowtie-1.0.1-linux-x86_64.zip/download
    curl -o samtools.tar.gz -L http://sourceforge.net/projects/samtools/files/latest/download?source=files
    wget ftp://ftp.genome.umd.edu/pub/jellyfish/jellyfish-2.1.3.tar.gz
    tar -xvf trinity.tar.gz
    rm trinity.tar.gz
    mv trinity* trinity
    unzip bowtie.zip
    tar -xvf samtools.tar.gz
    tar -xvf jellyfish-2.1.3.tar.gz
    #make samtools, jellyfish, trinity

 1. Using jellyfish version 2.1.3 instead of the the older version given in the trinity package. The new version will automatically merge the multiple files into one.
 
         #Use the -C flag for unstranded reads
         /mlig_working/software/jellyfish-2.1.3/bin/jellyfish count -t 32 -m 25 -s 10000000 -C /mlig_working/mlig.both.fa -o /mlig_working/mlig.both.jf
         /mlig_working/software/jellyfish-2.1.3/bin/jellyfish dump /mlig_working/mlig.both.jf > /mlig_working/mlig.both.kmers.fa
         
 2. Make a new AWS EBS/instance and run inchworm.
 
         /mlig_working/software/trinity/Inchworm/bin/inchworm --kmers /mlig_working/mlig.both.kmers.fa --run_inchworm -K 25 -L 25 --monitor 1 --num_threads 32 --DS > mlig.inchworm.K25.L25.DS.fa
 
 3. Perform jaccard clipping on the inchworm output.
     
         /mlig_working/software/trinity/util/support_scripts/inchworm_transcript_splitter.pl --iworm /mlig_working/mlig.inchworm.K25.L25.DS.fa --left /mlig_working/mlig.left.fastq --right /mlig_working/mlig.right.fastq --seqType fq --CPU 32
 
 4. Run Chrysalis on inchworm output. Make sure to have enough disk space for bowtie output in this step.
 
         #first set the stacksize to unlimited
         ulimit -u unlimited
         /mlig_working/software/trinity/Chrysalis/Chrysalis -i /mlig_working/mlig.both.fa \
                                          -iworm /mlig_working/mlig.inchworm.K25.L25.DS.fa.clipped.fa \
                                          -o /mlig_working/chrysalis -cpu 32 \
                                          -paired -reads_for_pairs /mlig_working/mlig.both.fa \
                                          -butterfly /mlig_working/software/trinity/Butterfly/Butterfly.jar \
                                          -max_mem_reads 10000000
                                          
 5. Run partition graph script to create component files
 
         /mlig_working/software/trinity/util/support_scripts/partition_chrysalis_graphs_n_reads.pl \
                                          --deBruijns /mlig_working/chrysalis/bundled_iworm_contigs.fasta.deBruijn \
                                          --componentReads /mlig_working/chrysalis/readsToComponents.out.sort -N 1000 -L 200 \
                                          --compdir chrysalis/Component_bins

 6. Run quantify graph program on partitioned components

 7. Run Butterfly on each component. '-F' is the max fragment length. '-L' is the minimum length of a transcript to be reported. '--path_reinforcement_distance' is the length of overlap needed for contiguous reads. Since we are using single end reads trimmed to maximum 40 bases, this distance should be 40 or less. 

```python
HTML(textarea('Script for generating buttefly commands','''
import sys,os

inFile = open(sys.argv[1],'r')

for line in inFile:
    path = line.strip().split()[1]
    print "java -Xmx7G -Xms1G -jar /mlig_working/software/trinity/Butterfly/Butterfly.jar -N 200000 -L 200 -F 400 -C " + path + ".graph --triplet-lock --generate-full-sequence-graphs --EM_REDUCE --path_reinforcement_distance=30"
'''))
```
