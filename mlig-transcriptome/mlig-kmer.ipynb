# K-mer analysis of raw M. Lignano sequence data

The goal is to produce a rarefaction plot of increasing input raw reads vs number of distinct k-mers. All pair-end reads are considered as 2 independent reads. Pair-end distance information doesn't matter in this case since we only care about the information content. 

The raw reads were consolidated into one giant .fasta file. This .fasta file is then shuffled with the script below. 

The script works by:

 1. Assigning a random number between 300-500 and create that number of empty files.
 
 2. Go through the raw .fasta file and distribute each read randomly to one of the bins.
 
 3. Concatenate the bins into a shuffled .fasta file
 
 4. Repeat this procedure several times.
 
In the case of the M. Lignano reads, we repeated this shuffling 5 times. This method of shuffling does not guarantee randomness. It will introduce some systematic bias towards the order of the reads in the original .fasta file. Alternate software is Alex Reynold's "sample" software: https://github.com/alexpreynolds/sample. However, this was taking too long (1 gig a day). 

```python
import sys, os
#import custom functions for displaying tables, bash commands
sys.path.append(os.path.abspath("/home/damian/"))
from dk_ipython import *

#Shuffle the raw sequence fasta file with this script. Not the best way to do this. Will still have some systematic bias.
#Shuffled with 5 rounds
HTML(textarea('Shuffle script','''import random, os, sys

def reShuffle(filePath, outDir, cycles = 0):
    binNum = random.randint(300,500)
    faFile = open(filePath)

    binFiles = {}
    for i in range(binNum):
        name = 'bin.' + str(i + 1).zfill(3) + '.fa'
        binFiles[i + 1] = open(outDir + name,'w')

    for line in faFile:
        stream = binFiles[random.randint(1,binNum)]
        stream.write(line)
        stream.write(faFile.next())

    faFile.close()
    for stream in binFiles.values():
        stream.close()

    os.system('cat ' + outDir + 'bin*.fa > ' + outDir + 'shuffled.fa')
    os.system('rm -f ' + outDir + 'bin*.fa')

    if cycles == 0:
        print 'Finished'
    else:
        print 'Finished, ' + str(cycles) + ' cycles remaining'
        reShuffle(outDir + 'shuffled.fa', outDir, cycles - 1)
    
reShuffle(sys.argv[1],sys.argv[2],int(sys.argv[3]))'''))
```

## K-mer counting using Jellyfish

Jellyfish 2.1.3 was used for k-mer counting. This process works by:

 1. Bin the shuffled .fasta file into bins of 50 million reads each.
 
 2. Create an empty file.
 
 3. Append one 50 million bin to the empty file.
 
 4. Run jellyfish to this initial 50 million reads file and produce a jellyfish db file for use later.
 
 5. Appending another 50 million bin to the reads file.
 
 6. Run jellyfish to this 100 million reads file and produce a jellyfish db file for use later.
 
 7. Repeat until all bins are used up.

## Read subset vs number of distinct k-mers analysis

In a normal genomic k-mer abundance graph, there is usually a peak at low k-mer abundances representing noise and a second peak further down which represents the genome coverage. As more genomic reads are added, the first peak might increase due to addition of noise and the second peak should shift to the right representing higher coverage. 

In a transcriptome k-mer abundance graph, we can imagine each transcript as a "genome". Each transcript should produe a k-mer abundance graph similar to a genome k-mer abundance graph. A transcriptome's k-mer abundance graph is then a mixture of these transcripts k-mer abundances. At extremely high coverage, we might be able to see the distribution mixture also shift to the right, resulting in a trough between the peak of noise at low k-mer abundance and the start of the mixture distribution. 

```python
import pandas as pd
import seaborn as sns
%matplotlib inline
from pandas.stats.api import ols
import matplotlib.pyplot as plt

#get k-mer information
histoFiles = [x for x in os.listdir('/home/share/projects/mlig_transcriptome/k-mer/') if x.split('.')[-1] == 'histo']

kmerData = []
mils = []
prefix = '/home/share/projects/mlig_transcriptome/k-mer/'
for histoFile in histoFiles:
    num = int(histoFile.split('.')[1]) + 1
    mil = num * 50
    mils.append(mil)
    histoData = [(int(x.split()[0]),int(x.split()[1])) for x in open(prefix + histoFile).read().strip().split('\n')[:-1]]
    kmerData.append(pd.Series([x[1] for x in histoData],index=[x[0] for x in histoData]))
kmerDF = pd.concat(kmerData,axis=1)
kmerDF.columns = mils
kmerDF = kmerDF.reindex_axis(sorted(kmerDF.columns), axis=1)
```

```python
sns.set_palette("Set2",10)
kmerDF.plot(logy=True, figsize=[15,10], linewidth=1)
```
