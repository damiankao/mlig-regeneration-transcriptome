## M. Lignano mapping

The transcriptome fasta file was given as `all.singlets.contigs.fa`. The file name is renamed to `tMl.1.3.fa`. The fasta headers for each sequence was renamed as `tMl_1.3_X`.

```python
count = 0
index = {}
newFa = open('tMl.1.3.renamed.fa','w')
for line in open('/home/share/projects/macregen/mapping/tMl.1.3.fa','r'):
    if line[0] == '>':
        oldName = line[1:].strip()
        newName = 'tMl_1.3_' + str(count).zfill(6)
        count += 1
        index[newName] = oldName
        newFa.write('>' + newName + "\n")
    else:
        newFa.write(line.strip() + "\n")
        
outFile = open('tMl.1.3.index','w')
outFile.write('\n'.join(['\t'.join(x) for x in index.items()]))
outFile.close()
newFa.close()
```

### Transcriptome stats

```python
from Bio import SeqIO
import numpy as np
def thou(num):
    return '{0:,}'.format(num)

faFile = open('tMl.1.3.renamed.fa','r')
data = {}
for record in SeqIO.parse(faFile,'fasta'):
    data[record.id] = str(record.seq)
```

```python
print 'Number of transcripts:', thou(len(data))
lengths = [len(v) for k,v in data.items()]
lengths.sort(reverse = True)
print 'Total base pairs:', thou(sum(lengths))
print 'Average length:', thou(np.mean(lengths))
print 'Minimum length:', thou(min(lengths))
print 'Maximum length:', thou(max(lengths))

total = 0
n50 = 0
half = sum(lengths) / 2
med = 0
for l in lengths:
    if total > half:
        n50 = l
        break
    med += 1
    total += l

print 'N50 length:', thou(l)
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")

from scipy.stats import gaussian_kde
density = gaussian_kde(lengths)
xs = np.linspace(min(lengths),max(lengths),100)
density.covariance_factor = lambda : .1
density._compute_covariance()

plt.figure(figsize=[14,5])
plt.title('Transcript length histogram')
plt.plot(xs,density(xs))
xs = np.linspace(min(lengths),l,100)

plt.figure(figsize=[14,5])
plt.title('Transcript length histogram from mininum to N50 length')
plt.plot(xs,density(xs))
```

## Cleaning raw read data

```python
import os

inFiles = [f for f in os.listdir('/home/share/projects/macregen/raw') if f.split('.')[-1] == 'gz']

cmds = []
for f in inFiles:
    name = '.'.join(f.split('.')[:-2])
    cmd = 'java -jar /home/share/software/Trimmomatic-0.32/trimmomatic-0.32.jar SE -threads 32 -phred33 ' + f + ' ' + name + '.trimmed.fastq.gz LEADING:20 TRAILING:20 MINLEN:40'
    cmds.append(cmd)
print '\n'.join(cmds)
```
