# Preping reads for assembly

From all the reads, produce 4 sets of files from trimmed reads:

 - PE, stranded libraries (left.fastq, right.fastq)
 - PE, long insert, non-stranded libraries (left.long.fastq, right.long.fastq)
 - SE, non-stranded libaries (all.se.fastq)
 - All libraries represented as a single reads in a fasta file for inchworm k-mer saturation analysis

```python
import gzip, os, sys
#import custom functions for displaying tables, bash commands
sys.path.append(os.path.abspath("/home/damian/"))
from dk_ipython import *
from IPython.display import HTML
import json
```

```python
import json
trimmLog = open('resource/trimPE.err')
libs = []
for line in trimmLog:
    if line.find('Started with arguments') != -1:
        #exclude the unstranded A1, A2, A1350, A2340 libraries
        if line.find('_A') == -1:
            libs.append(line.strip().split('.fastq.gz')[0].split()[-1].split('/')[-1][:-1])

prefix = '/mlig_raw/trim/'
f = [prefix + k + '1.fastq.gz.paired.trimmed.fastq.gz' for k in libs]
r = [prefix + k + '2.fastq.gz.paired.trimmed.fastq.gz' for k in libs]

HTML(textarea('Script for PE, stranded fastq.gz','''import gzip
pairA = ''' + json.dumps(f) + '''
pairB = ''' + json.dumps(r) + '''

outA = open("left.stranded.fastq",'w')
outB = open("right.stranded.fastq",'w')

aCount = 0
for a in pairA:
    name = a.split('/')[-1].split('.')[0][:-3]

    stream = gzip.open(a,'rb')
    for line in stream:
        seq = stream.next()
        stream.next()
        qual = stream.next()
        outA.write('@r' + str(aCount).zfill(10) + '_' + name + '/1\\n' + seq + '+\\n' + qual)
        aCount += 1

bCount = 0
for b in pairB:
    name = b.split('/')[-1].split('.')[0][:-3]

    stream = gzip.open(b,'rb')
    for line in stream:
        seq = stream.next()
        stream.next()
        qual = stream.next()
        outB.write('@r' + str(bCount).zfill(10) + '_' + name + '/2\\n' + seq + '+\\n' + qual)
        bCount += 1

outA.close()
outB.close()'''))
```

```python
import json
libs = []
trimmLog = open('resource/trimSE.err')
for line in trimmLog:
    if line.find('Started with arguments') != -1:
        libs.append(prefix + line.strip().split('.fastq.gz')[0].split()[-1].split('/')[-1] + '.trimmed.fastq.gz')
trimmLog.close()

HTML(textarea('Script for all SE reads','''import gzip
allFiles = ''' + json.dumps(libs) + '''

outA = open("all.se.fastq",'w')

aCount = 0
for a in allFiles:
    name = a.split('/')[-1].split('.')[0][:-3]

    stream = gzip.open(a,'rb')
    for line in stream:
        seq = stream.next()
        stream.next()
        qual = stream.next()
        outA.write('@r' + str(aCount).zfill(10) + '_' + name + '\\n' + seq + '+\\n' + qual)
        aCount += 1

outA.close()'''))
```

```python
import json
trimmLog = open('resource/trimPE.err')
libs = []
prefix = '/mlig_raw/trim/'
for line in trimmLog:
    if line.find('Started with arguments') != -1:
        libs.append(prefix + line.strip().split('.fastq.gz')[0].split()[-1].split('/')[-1][:-1] + '1.fastq.gz.paired.trimmed.fastq.gz')
        libs.append(prefix + line.strip().split('.fastq.gz')[0].split()[-1].split('/')[-1][:-1] + '2.fastq.gz.paired.trimmed.fastq.gz')
        libs.append(prefix + line.strip().split('.fastq.gz')[0].split()[-1].split('/')[-1][:-1] + '1.fastq.gz.unpaired.trimmed.fastq.gz')
        libs.append(prefix + line.strip().split('.fastq.gz')[0].split()[-1].split('/')[-1][:-1] + '2.fastq.gz.unpaired.trimmed.fastq.gz')
trimmLog.close()
trimmLog = open('resource/trimSE.err')
for line in trimmLog:
    if line.find('Started with arguments') != -1:
        libs.append(prefix + line.strip().split('.fastq.gz')[0].split()[-1].split('/')[-1] + '.trimmed.fastq.gz')
trimmLog.close()

HTML(textarea('Script for consolidating all libraries as single reads for inchworm k-mer analysis','''import gzip
allFiles = ''' + json.dumps(libs) + '''

outA = open("all.fasta",'w')

aCount = 0
for a in allFiles:
    name = a.split('/')[-1].split('.')[0][:-3]

    stream = gzip.open(a,'rb')
    for line in stream:
        seq = stream.next()
        stream.next()
        qual = stream.next()
        outA.write('>r' + str(aCount).zfill(10) + '_' + name + '\\n' + seq)
        aCount += 1

outA.close()'''))
```
