### Assembly metrics

```python
prefix = '/home/share/projects/mlig_transcriptome/oases/assemblies/'

fs = os.listdir(prefix)
from collections import defaultdict
lengths = defaultdict(list)
for f in fs:
    inFile = open(prefix + f)
    k = f.split('.')[0]
    for line in inFile:
        if line[0] == '>':
            l = int(line.strip().split('_')[-1])
            lengths[k].append(l)
    inFile.close()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

lengthItems = lengths.items()
lengthItems.sort(key = lambda x : x[0])

def n50(ls):
    ls.sort(reverse = True)
    half = sum(ls) / 2
    total = 0
    for l in ls:
        total += l
        if total > half:
            return l
    
metricTable = ListTable()
metricTable.append(['k-value','transcripts #','mean length','median length','N50 length'])
for k, l in lengthItems:
    s = pd.Series(l)
    metricTable.append([k,commas(len(s)),s.mean(),s.median(),n50(l)])
metricTable
```

```python
numCol = 2
numRow = 5
fig, axes = plt.subplots(ncols = numCol, nrows = numRow, figsize=[20, 20])
i = 0
for k, l in lengthItems:
    r = i / numRow
    c = i % numRow
    l = pd.Series(l)
    l[l < 1000].hist(ax = axes[c,r], bins=100)
    axes[c,r].set_title(k)
    axes[c,r].set_xlim(100, 1000)
    axes[c,r].set_ylim(0,100000)
    i += 1
fig.tight_layout()
    
```
