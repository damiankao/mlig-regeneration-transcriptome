# M. Lignano transcriptome assembly stats

The transcriptome was assembled with Trinity. 

```python
import sys,os
from collections import defaultdict
from Bio import SeqIO
#import custom functions for displaying tables, bash commands
sys.path.append(os.path.abspath("/home/damian/"))
from dk_ipython import *
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

faFile = open('/home/share/projects/mlig_transcriptome/assembly_stranded/mlig.trinity.fasta')
header = []
lengths = []
for line in faFile:
    if line[0] == '>':
        header.append(line[1:].split()[0])
        lengths.append(int(line.split('len=')[-1].split()[0]))
```

```python
lengths = pd.Series(lengths, index = header)
print 'Number of transcripts assembled:', commas(len(header))
print 'Number of Trinity defined genes:', commas(len(set(['_'.join(item.split('_')[:2]) for item in header])))
print 'Mean transcript length:', lengths.mean()
print 'Median transcript length:', lengths.median()
lengths.hist(bins=range(0,4000,20), figsize=[15,6])
```

### Use transdecoder to determine coding potential
Run TransDecoder:

    TransDecoder -t /home/share/projects/mlig_transcriptome/assembly/mlig.trinity.fasta --workdir /home/share/projects/mlig_transcriptome/assembly/transdecoder --search_pfam /home/share/software/TransDecoder_r20140704/pfam/Pfam-AB.hmm.bin --CPU 25

```python
from collections import Counter

mrnaFile = open('/home/share/projects/mlig_transcriptome/assembly_stranded/mlig.trinity.fasta.transdecoder.mRNA')

mrnaHeader = []
mrnaLengths = []
transType = {}
for record in SeqIO.parse(mrnaFile,'fasta'):
    tid = record.description.split()[0]
    mrnaHeader.append(tid)
    mrnaLengths.append(len(str(record.seq)))
    transType[tid.split('|')[0]] = record.description.split('type:')[-1].split()[0]
```

```python
mrnaCount = Counter([k.split('|')[0] for k in mrnaHeader])
print 'Multiple mRNAs can be found in one single Trinity defined transcript (chimeric).'
print 'Number of transdecoder defined mRNA:', commas(len(mrnaHeader))
print
print 'Number of Trinity transcripts with transdecoder mRNA:', commas(len(set([x.split('|')[0] for x in mrnaHeader])))
print 'Number of Trinity genes with transdecoder mRNA:', commas(len(set(['_'.join(x.split('|')[0].split('_')[:2]) for x in mrnaHeader])))
print
print 'Number of Trinity transcripts with multiple transdecoder mRNA:', commas(len([k for k,v in mrnaCount.items() if v > 1]))
print 'Number of Trinity genes with multiple transdecoder mRNA:', commas(len(set(['_'.join(k.split('|')[0].split('_')[:2]) for k,v in mrnaCount.items() if v > 1])))
```

```python
lengths.hist(bins=range(0,4000,20), figsize=[15,6])
pd.Series([lengths[x] for x in set([x.split('|')[0] for x in mrnaHeader])]).hist(bins=range(0,4000,20), figsize=[15,6])
plt.legend(['all Trinity transcripts','all Trinity transcripts with \nat least one transdecoder mRNA'])
```

```python
mrnaHeader_single = [k for k,v in mrnaCount.items() if v == 1]
print 'Number of Trinity transcripts with one transdecoder mRNA:', commas(len(mrnaHeader_single))
transType_one_counter = Counter([transType[tid] for tid in mrnaHeader_single])
for t, v in transType_one_counter.items():
    print '   ' + t + ': ' + commas(v)
print
print "*5' partial means missing 5' end"
print "*3' partial means missing 3' end"
```

```python
pd.Series([lengths[x] for x in [item.split('|')[0] for item in mrnaHeader]]).hist(bins=range(0,4000,30), figsize=[20,8], alpha=0.1)
pd.Series([lengths[x] for x in [item.split('|')[0] for item in mrnaHeader_single]]).hist(bins=range(0,4000,30), figsize=[20,8], alpha=0.1)
for t in transType_one_counter.keys():
    tids = [k for k in mrnaHeader_single if transType[k] == t]
    lengths[tids].hist(bins=range(0,4000,30), figsize=[15,6], alpha=0.6)
plt.title('ORF completeness of Trinity transcripts with one transdecoder mRNA')
plt.legend(['all Trinity transcripts with \nat least one transdecoder mRNA','all Trinity transcripts \nwith one transdecoder mRNA'] +transType_one_counter.keys())
```

### Blast to Uniprot/Trembl metazoa

    cat mlig.trinity.fasta | parallel --pipe -k -j 16 --block 10k --recstart ">" "blastx -db /home/share/blastdb/uniprot_trembl/metazoa -query - -outfmt '6 std qlen slen qframe stitle'" > mlig.trinity.blast

```python
blastFile = open('/home/share/projects/mlig_transcriptome/assembly_stranded/blast/mlig.trinity.blast')
filterFile = open('/home/share/projects/mlig_transcriptome/assembly_stranded/blast/mlig.trinity.filtered.blast','w')

used = {}
#filter out evalue > 1e-5, subject hit = 'predicted protein', redundant hits
for line in blastFile:
    meta = line.strip().split('\t')
    if not used.has_key(meta[0] + meta[1]):
        if float(meta[10]) < 1e-5:
            if meta[-1].lower().find('predicted protein') == -1:
                filterFile.write(line)
        used[meta[0] + meta[1]] = True

filterFile.close()
blastFile.close()
```

```python
from collections import defaultdict
filteredFile = open('/home/share/projects/mlig_transcriptome/assembly_stranded/blast/mlig.trinity.filtered.blast')
topFile = open('/home/share/projects/mlig_transcriptome/assembly_stranded/blast/mlig.trinity.top.blast','w')

#get top 5 blast hits with query, subeject, evalue, frame
tid = defaultdict(int)
for line in filteredFile:
    meta = line.strip().split('\t')
    tid[meta[0]] += 1
    if tid[meta[0]] < 6:
        topFile.write('\t'.join([meta[0],meta[-1],meta[10],meta[12]]) + '\n')

topFile.close()
filteredFile.close()
```

```python
td_tids = []
for line in open('/home/share/projects/mlig_transcriptome/assembly_stranded/mlig.trinity.fasta.transdecoder.mRNA'):
    if line[0] == '>':
        td_tids.append(line.split('|')[0][1:])
        
annoFile = open('/home/share/projects/mlig_transcriptome/assembly_stranded/blast/mlig.trinity.top.blast')

anno = []
for line in annoFile:
    meta = line.strip().split('\t')
    anno.append(meta[0])
anno = list(set(anno))
```

```python
print 'Number of transcript with coding potential detected with transdecoder:', commas(len(td_tids))
print 'Number of transcripts with blast annotations < 1E-5:', commas(len(anno))
print 'Intersection:', commas(len(set(set(td_tids) & set(anno))))
print 'Union:', commas(len(set(set(td_tids) | set(anno))))

print commas(len(set(['_'.join(x.split('_')[:2]) for x in list(set(set(td_tids) | set(anno)))])))
```
