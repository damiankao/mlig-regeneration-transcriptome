# Differential expression 

A transcript level and gene level DE will be performed. Two popular packages are used for isoform read distribution: RSEM and eXpress. EXpress is much easier to use.

However, RSEM/eXpress outputs decimal counts weighed by length and multi-mapping. DE packages like DESeq2 and EdgeR uses discrete count models for their DE analysis and doesn't take into account isoform abundances. The eXpress manual recommends just simply rounding the decimal counts to the nearest integer. There is also EBSeq which was designed for use with RSEM output and can take decimal counts. 

Three different DE analysis will be performed:

 1. eXpress data + EBSeq
 2. gene level counts (sum of all isoform's uniquely mapping reads defined by eXpress) + DESeq2
 3. rounded eXpress data + DESeq2
 
The goal is to get DE genes for control vs regen at each time-point.

## Correlation among replicates

```python
#produce gene level counts and rounded counts
counts = open('/home/share/projects/mlig_regen/express/mlig_regen.express.unique.count')
geneOut = open('/home/share/projects/mlig_regen/express/mlig_regen.express.gene.count','w')
header = counts.next()
geneOut.write(header)

genes = {}
for line in counts:
    meta = line.strip().split()
    tid = meta[0]
    gid = '_'.join(tid.split('_')[:2])
    if not genes.has_key(gid):
        genes[gid] = [int(x) for x in meta[1:]]
    else:
        for i, x in enumerate(meta[1:]):
            genes[gid][i] += int(x)
                
for gid, uniqueCount in genes.items():
    geneOut.write(gid + '\t' + '\t'.join([str(x) for x in uniqueCount]) + '\n')
    
counts.close()
geneOut.close()

expressCount = open('/home/share/projects/mlig_regen/express/mlig_regen.express.count')
roundedOut = open('/home/share/projects/mlig_regen/express/mlig_regen.express.rounded.count','w')
header = expressCount.next()
roundedOut.write(header)

for line in expressCount:
    meta = line.strip().split()
    tid = meta[0]

    #Using the native round() might not be the best way
    #Potential floating point error when number is very close to 0.5, ie round(1.4999999999999) = 2
    #But only if precision is high, so it should not affect these counts
    rounded = [int(round(float(x))) for x in meta[1:]]
    roundedOut.write(tid + '\t' + '\t'.join([str(x) for x in rounded]) + '\n')
    
expressCount.close()
roundedOut.close()
```

## Comparison of DE analysis

Differentially expressed transcripts at p-value > 0.05 in EBSeq seems to be inclusive of rounded DESeq DE transcripts. DESeq is more strict calling only ~10% of the EBSeq DE transcript as differentially expressed. I am going to use the intersection of both methods as the differentially expressed list. 

```python
deTable = ListTable()
rows = []
for i in set(time):
    rows.append([i,str(len(set(set(deseq_rounded_de[i]) & set(ebseq_de[i])))),str(len(deseq_rounded_de[i])), str(len(ebseq_de[i]))])
rows.sort(key = lambda x : x[0])
deTable.append(['Time','Intersection','Rounded DESeq','EBSeq'])
deTable.extend(rows)
deTable
```
