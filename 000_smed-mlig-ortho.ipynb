# S.mediterranea and M.lignano orthologs

All vs all tblastx done by:

    tblastx -query smed.mlig.fa -db smed.mlig -evalue 1e-5 -num_threads 30 -outfmt '6 qseqid sseqid qlen slen pident length qstart qend sstart send evalue' > smed.mlig.blast &
    
The resulting blast file was filtered for redundancy and self hits. Best HSPs were kept for each alignment.


```python
from collections import defaultdict
blastFile = open('/home/share/projects/mlig_transcriptome/stem/ortho/smed.mlig.uniq.filterSet.blast')
hits = defaultdict(list)

import math
for line in blastFile:
    meta = line.strip().split()
    meta[-1] = float(meta[-1])
    
    if meta[-1] < 1e-5:
        hits[meta[0]].append((meta[1],meta[-1]))
```

```python
def recipHit(ethresh):
    ox = defaultdict(list)
    for tid, hids in hits.items():
        if tid[0] == 'O':
            hids = [x for x in hids if x[1] < ethresh]
            hids.sort(key = lambda x : x[1])

            for i,hid in enumerate(hids):
                if hits.has_key(hid[0]):
                    recips = hits[hid[0]]
                    recips = [x for x in recips if x[1] < ethresh]
                    recips.sort(key = lambda x : x[1])

                    for j, recip in enumerate(recips):
                        if recip[0] == tid:
                            ox[tid].append((hid[0],i,j))
    return ox
                        
```

```python
d = recipHit(1e-15)
```

```python
recip2hits = defaultdict(list)
for tid, hitStats in d.items():
    for hitStat in hitStats:
        rank = max(hitStat[1],hitStat[2])
        if rank < 2:
            recip2hits[tid].append(hitStat[0])
```

```python
outFile = open('/home/share/projects/mlig_transcriptome/stem/ortho/ortho.index','w')
for tid, hids in recip2hits.items():
    outFile.write(tid + '\t' + ','.join(hids) + '\n')
```

```python
prefix = '/home/share/projects/mlig_transcriptome/stem/ortho/'
tailLessA = open(prefix + 'mlig_regen_reg_000h_1_1.count')
counts = {}
for line in tailLessA:
    meta = line.strip().split()
    counts[meta[0]] = float(meta[1])

tailLessB = open(prefix + 'mlig_regen_reg_000h_2_1.count')

for line in tailLessB:
    meta = line.strip().split()
    counts[meta[0]] += float(meta[1])
    
rank = counts.items()
rank.sort(key = lambda x : x[1], reverse = True)
rank = [x[0] for x in rank]
```

```python
neoblastAnno = open(prefix + 'neoblast.data')
knownAnno = open(prefix + 'known.anno')
total = sum(counts.values())

display = {}
for line in neoblastAnno:
    meta = line.strip().split('\t')
    display[meta[0]] = meta[1]
    
for line in knownAnno:
    meta = line.strip().split('\t')
    display[meta[0]] = meta[1]

displayTable = ListTable()
displayTable.append(['s.med ID','annotation','m.lig ID','% tags','avg tags','avg % rank'])
for tid, anno in display.items():
    if recip2hits.has_key(tid):
        recips = recip2hits[tid]
        c = []
        r = []
        for recip in recips:
            if counts.has_key(recip):
                c.append(counts[recip])
                r.append(rank.index(recip))
                
        if len(c) > 0:
            c = sum(c) / float(len(c))
            r = sum(r) / float(len(r))
        else:
            c = -1
            r = -1
                
        displayTable.append([tid,anno,','.join(recip2hits[tid]),str(c / total),str(c), str(100 * r / float(len(rank)))])
```

Following table shows known S.med genes + predicted S.med neoblast genes. The corresponding M.lig ID is given along with:

 - % tags = avg number of tags per m.lig ID divided by total tag count for the entire library
 - avg tags = average number of tags (if multiple m.lig IDs)
 - avg % rank = percentage ranking of expression (lower percent, the higher expressed)
