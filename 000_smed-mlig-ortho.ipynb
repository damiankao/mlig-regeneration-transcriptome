All vs all tblastx done by:

    tblastx -query smed.mlig.fa -db smed.mlig -evalue 1e-5 -num_threads 30 -outfmt '6 qseqid sseqid qlen slen pident length qstart qend sstart send evalue' > smed.mlig.blast &
    
The resulting blast file was filtered for redundancy and self hits. Best HSPs were kept for each alignment.


```python
from collections import defaultdict
blastFile = open('/home/share/projects/mlig_transcriptome/stem/ortho/smed.mlig.uniq.filterSet.blast')
hits = defaultdict(list)

import math
for line in blastFile:
    meta = line.strip().split()
    meta[-1] = float(meta[-1])
    
    if meta[-1] < 1e-5:
        hits[meta[0]].append((meta[1],meta[-1]))
```

```python
def recipHit(ethresh):
    ox = defaultdict(list)
    for tid, hids in hits.items():
        if tid[0] == 'O':
            hids = [x for x in hids if x[1] < ethresh]
            hids.sort(key = lambda x : x[1])

            for i,hid in enumerate(hids):
                if hits.has_key(hid[0]):
                    recips = hits[hid[0]]
                    recips = [x for x in recips if x[1] < ethresh]
                    recips.sort(key = lambda x : x[1])

                    for j, recip in enumerate(recips):
                        if recip[0] == tid:
                            ox[tid].append((hid[0],i,j))
    return ox
                        
```

```python
d = recipHit(1e-15)
```

```python
recip2hits = defaultdict(list)
for tid, hitStats in d.items():
    for hitStat in hitStats:
        rank = max(hitStat[1],hitStat[2])
        if rank < 2:
            recip2hits[tid].append(hitStat[0])
```

```python
outFile = open('/home/share/projects/mlig_transcriptome/stem/ortho/ortho.index','w')
for tid, hids in recip2hits.items():
    outFile.write(tid + '\t' + ','.join(hids) + '\n')
```
