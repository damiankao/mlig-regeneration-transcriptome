# QC of raw data with fastqc

Raw data is stored in `/home/share/ngsData/mlig`. Fastqc was ran with: 

    fastqc -o /home/share/ngsData/mlig/fastqc -t 30 *.gz
    
Fastqc results are stored in `/home/share/ngsData/mlig/fastqc`

```python
#import libraries and custom display libraries
import os, sys, gzip, base64
#import custom functions for displaying tables, bash commands
sys.path.append(os.path.abspath("/home/damian/"))
from dk_ipython import *
from IPython.display import HTML
from collections import defaultdict
```

```python
prefix = '/home/share/ngsData/mlig/fastqc/'
fastqc = os.listdir('/home/share/ngsData/mlig/fastqc')

conds = defaultdict(list)

for d in fastqc:
    name = '_'.join(d.split('_')[:-3])
    if d.find('pilot') != -1:
        name = '_'.join(d.split('_')[:-2])
    qual = prefix + d + '/Images/per_base_quality.png'
    conds[name].append(qual)

qualTable_pilot = ListTable()
qualTable_regen = ListTable()
qualTable_pilot.append(['Pilot PE Libraries'])
qualTable_regen.append(['Regeneration SE Library'])
            
for cond, paths in conds.items():
    row = [cond]
    paths.sort()
    for path in paths:    
        row.append(embedImage(path))
            
    if cond.find('pilot') != -1:
        qualTable_pilot.append(row)
    else:
        qualTable_regen.append(row)
        
qualTable_pilot.sort(key = lambda x : x[0])
qualTable_regen.sort(key = lambda x : x[0])
qualTable_pilot
```
