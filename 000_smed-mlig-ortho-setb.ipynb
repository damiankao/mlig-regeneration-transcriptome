```python
from collections import defaultdict
blastFile = open('/home/share/projects/mlig_transcriptome/stem/ortho/smed.mlig.uniq.filterSet.blast')
hits = defaultdict(list)

import math
for line in blastFile:
    meta = line.strip().split()
    meta[-1] = float(meta[-1])
    
    if meta[-1] < 1e-5:
        hits[meta[0]].append((meta[1],meta[-1]))
        
def recipHit(ethresh):
    ox = defaultdict(list)
    for tid, hids in hits.items():
        if tid[0] == 'O':
            hids = [x for x in hids if x[1] < ethresh]
            hids.sort(key = lambda x : x[1])

            for i,hid in enumerate(hids):
                if hits.has_key(hid[0]):
                    recips = hits[hid[0]]
                    recips = [x for x in recips if x[1] < ethresh]
                    recips.sort(key = lambda x : x[1])

                    for j, recip in enumerate(recips):
                        if recip[0] == tid:
                            ox[tid].append((hid[0],i,j))
    return ox
                        
d = recipHit(1e-15)
recip2hits = defaultdict(list)
for tid, hitStats in d.items():
    for hitStat in hitStats:
        rank = max(hitStat[1],hitStat[2])
        if rank < 2:
            recip2hits[tid].append(hitStat)
            
allTids = []
for k,vs in recip2hits.items():
    allTids.append(k)
    allTids.extend([x[0] for x in vs])

allTids = set(allTids)

faFile = open('/home/share/projects/mlig_transcriptome/stem/ortho/smed.mlig.fa')
from Bio import SeqIO
allSeqs = {}
for record in SeqIO.parse(faFile,'fasta'):
    if record.id in allTids:
        allSeqs[record.id] = str(record.seq)
        
oxSeqs = {}
faFile = open('/home/share/transcriptome/smed/ox_smed_v1/ox_Smed_v1.fa')
for record in SeqIO.parse(faFile,'fasta'):
    if record.id in recip2hits.keys():
        oxSeqs[record.id] = str(record.seq)
```

```python
from Bio.Seq import Seq

def getMaxFrag(prot, forceStart = False):
    inds = [0] + [x + 1 for x, c in enumerate(str(prot)) if c == '*'] + [len(prot)]
    frags = [[inds[x],inds[x + 1], prot[inds[x]:inds[x + 1]]] for x in range(len(inds) - 1)]
    filterFrags = []
    
    for i in range(len(frags)):
        mCheck = False
        if str(frags[i][2]).find('M') != -1:
            mCheck = True
            frags[i][0] += str(frags[i][2]).index('M')
        
        if not forceStart:
            if i == 0:
                filterFrags.append(frags[i])
        else:
            if mCheck:
                filterFrags.append(frags[i])
                
    return filterFrags
    
def longestORF(seq, forceForward = False, forceStart = False):
    seqLen = len(seq)
    seq = Seq(seq)
    revSeq = seq.reverse_complement()
    frames = []
    for i in range(3):
        tr = seq[i:].translate()
        revTr = revSeq[i:].translate()
        
        forward_frags = getMaxFrag(tr, forceStart)
        reverse_frags = getMaxFrag(revTr, forceStart)

        if len(forward_frags) > 0:
            maxForward = max(forward_frags,key = lambda x : x[1] - x[0])
            frames.append((maxForward,(i + 1)))
        
        if not forceForward:
            if len(reverse_frags) > 0:
                maxReverse = max(reverse_frags,key = lambda x : x[1] - x[0])
                frames.append((maxReverse, (-i - 1)))

    maxFrame = max(frames, key = lambda x : x[0][1] - x[0][0])
    maxCoords = maxFrame[0]
    frame = maxFrame[1]

    if frame > 0:
        return ((maxCoords[0] * 3) + frame - 1, (maxCoords[1] * 3) + frame - 1)
    else:
        return (seqLen + frame + 1 - (maxCoords[0] * 3), seqLen + frame + 1 - (maxCoords[1] * 3))
```

```python
def primer3(seq, orfStart, orfEnd):
    header = '''SEQUENCE_ID=__NAME__
SEQUENCE_TEMPLATE=__SEQ__
SEQUENCE_PRIMER_PAIR_OK_REGION_LIST=__REGION__
PRIMER_TASK=pick_pcr_primers
PRIMER_PICK_LEFT_PRIMER=1
PRIMER_PICK_INTERNAL_OLIGO=0
PRIMER_PICK_RIGHT_PRIMER=1
PRIMER_PRODUCT_SIZE_RANGE=400-1000
PRIMER_PRODUCT_OPT_SIZE=900
PRIMER_OPT_SIZE=24
PRIMER_MIN_SIZE=18
PRIMER_MAX_SIZE=27
PRIMER_MIN_GC=20.0
PRIMER_MAX_GC=80.0
PRIMER_MIN_TM=57.0
PRIMER_MAX_TM=68.0
PRIMER_EXPLAIN_FLAG=0
='''
    cmd = 'primer3_core temp.p3 > temp.out'
    
    endBound = len(seq) - 5
    regionA = '5,' + str(orfStart - 5)
    regionB = str(orfStart + 5) + ',' + str(orfEnd - orfStart - 5)
    regionC = str(min(endBound, orfEnd + 5)) + ',' + str(endBound - min(endBound, orfEnd + 5))
    regionAll = '5,' + str(endBound - 5)
    
    if orfEnd + 5 > endBound:
        regionString = regionA + ',' + regionB + ';' + regionAll + ',' + regionAll
    else:
        regionString = regionA + ',' + regionB + ';' + regionB + ',' + regionC + ';' + regionAll + ',' + regionAll
    
    tmpFile = open('temp.p3','w')
    tmpFile.write(header.replace('__NAME__','sequence')\
                  .replace('__SEQ__',str(seq))\
                  .replace('__REGION__',regionString))
    tmpFile.close()
    os.system(cmd)
    result = dict([x.split('=') for x in open('temp.out','r').read().strip().split('\n')])
    
    longest = max([[i,result['PRIMER_PAIR_' + str(i) + '_PRODUCT_SIZE']] \
               for i in range(int(result['PRIMER_PAIR_NUM_RETURNED']))],\
        key = lambda x : x[1])[0]
        
    
    htmlString = ''
    if int(result['PRIMER_PAIR_NUM_RETURNED']) > 0:
        htmlString += 'product size: ' + result['PRIMER_PAIR_' + str(longest) + '_PRODUCT_SIZE'] + '<BR>'
        htmlString += 'primer left: ' + result['PRIMER_LEFT_' + str(longest) + '_SEQUENCE'] + '<BR>'
        htmlString += 'primer left GC: ' + result['PRIMER_LEFT_' + str(longest) + '_GC_PERCENT'] + '<BR>'
        htmlString += 'primer left TM: ' + result['PRIMER_LEFT_' + str(longest) + '_TM'] + '<BR>'
        htmlString += 'primer right: ' + result['PRIMER_RIGHT_' + str(longest) + '_SEQUENCE'] + '<BR>'
        htmlString += 'primer right GC: ' + result['PRIMER_RIGHT_' + str(longest) + '_GC_PERCENT'] + '<BR>'
        htmlString += 'primer right TM: ' + result['PRIMER_RIGHT_' + str(longest) + '_TM'] + '<BR>'
 
        return (result['PRIMER_LEFT_' + str(longest) + '_SEQUENCE'],\
                result['PRIMER_RIGHT_' + str(longest) + '_SEQUENCE'],\
                result['PRIMER_PAIR_' + str(longest) + '_PRODUCT_SIZE'],\
                htmlString)
```
