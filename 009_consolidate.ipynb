# Consolidated assemblies

 1. Run CD-HIT-EST on assemblies grouped by assembly method (oases, trinity, transabyss...)
         cat trinity.*.fa > trinity.x.fa
         cd-hit-est -i trinity.x.fa -o trinity.x.cdhit.fa -M 0 -T 32 -c 0.98 -d 0

 2. Map reads uniquely back to CD-HIT-EST assemblies
 3. Reads that map to multiple assembly methods are recorded
 4. Cluster base on multi-method-mapped reads

### CD-HIT-EST and map

```python
fs = [x for x in os.listdir('/home/share/ngsData/mlig/') if x.split('.')[-1] == 'gz']
template = 'bowtie2 --very-fast --no-unal -p 32 -x EBWT -U FASTQ | samtools view -bS - > '
cmds = ''

for f in fs:
    f = '/mlig/trim/' + f + '.paired.trimmed.fastq.gz'
    cmds += template.replace('EBWT','/mlig/bt/oases').replace('FASTQ',f) + f.split('/')[-1].split('.')[0] + '.oases.bam\n'
    cmds += template.replace('EBWT','/mlig/bt/trinity').replace('FASTQ',f) + f.split('/')[-1].split('.')[0] + '.trinity.bam\n'
    cmds += template.replace('EBWT','/mlig/bt/soap').replace('FASTQ',f) + f.split('/')[-1].split('.')[0] + '.soap.bam\n'
    cmds += template.replace('EBWT','/mlig/bt/reassembly').replace('FASTQ',f) + f.split('/')[-1].split('.')[0] + '.reassembly.bam\n'
    cmds += template.replace('EBWT','/mlig/bt/transabyss').replace('FASTQ',f) + f.split('/')[-1].split('.')[0] + '.transabyss.bam\n'
    
HTML(textarea('Bowtie2 mapping commands',cmds))
```

```python
from collections import defaultdict

prefix = '/home/share/projects/mlig_transcriptome/read_cluster/'

fs = list([x for x in os.listdir(prefix) if x.split('.')[-1] == 'data'])

groups = defaultdict(list)
for f in fs:
    lib = f.split('.')[0]
    groups[lib].append(f)

def getReadIndex(mapFiles):
    reads = defaultdict(list)
    for mapFile in mapFiles:
        inFile = open(prefix + mapFile)
        for line in inFile:
            meta = line.strip().split()
            reads[meta[0]].append(meta[1])
            
    return reads
```
