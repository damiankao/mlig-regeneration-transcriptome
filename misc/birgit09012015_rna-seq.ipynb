# Absolute tag counts and diffferential expression analysis in RNA-seq

Differential expression was done with a statistical model that accounts for the number of tag counts. 

The number of tag counts per gene of sample is essentially modeled with a Poisson distribution. This is not strictly true, but it's fine to think about it as a poisson for now.

Here is a Poisson distribution for a tag count of 10:

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy.stats
from scipy.stats import poisson
from scipy.stats import gamma
%matplotlib inline
```

```python
k = np.arange(25)
colors = matplotlib.rcParams['axes.color_cycle'] 
plt.figure(figsize=(12,4))
for i, lambda_ in enumerate([10]):
    plt.bar(k, poisson.pmf(k, lambda_), label=lambda_, color=colors[i], alpha=0.4, edgecolor=colors[i], lw=3)
    plt.legend()
plt.title("Poisson distribution")
plt.xlabel("$count$")
plt.ylabel("PDF at count")
```

Each bar shows the probability of observing a specific count.

Now let's look at counts of two different genes at counts of 10 and 20:

```python
print 'Poisson distribution of tag counts of 10 and 20'
k = np.arange(40)
colors = matplotlib.rcParams['axes.color_cycle'] 
plt.figure(figsize=(12,4))
for i, lambda_ in enumerate([10, 20]):
    plt.bar(k, poisson.pmf(k, lambda_), label=lambda_, color=colors[i], alpha=0.4, edgecolor=colors[i], lw=3)
    plt.legend()
plt.title("Poisson distribution")
plt.xlabel("$k$")
plt.ylabel("PDF at k")
```

Notice how their distributions overlap. This means that there is a chance that these two genes are expressed at the same level (not differentially expressed).

Now let's look at the counts of two different genes at counts of 100 and 200:

```python
print 'Poisson distribution of tag counts of 100 and 200'
k = np.arange(260)
colors = matplotlib.rcParams['axes.color_cycle'] 
plt.figure(figsize=(12,4))
for i, lambda_ in enumerate([100, 200]):
    plt.bar(k, poisson.pmf(k, lambda_), label=lambda_, color=colors[i], alpha=0.4, edgecolor=colors[i], lw=3)
    plt.legend()
plt.title("Poisson distribution")
plt.xlabel("$k$")
plt.ylabel("PDF at k")
```

Notice now that the two distributions don't overlap signifying that they are most likely differentially expressed. Although, the fold change is the same (fc of 2), the higher counts allowed a better differentiation of the variances of the two distributions.

# Negative binomial

The tag counts are actually modeled by a negative binomial distribution which is a mixture of gamma and poisson distribution. Think of it like this:

 - You have multiple biological replicates of a sample, the variance of these biological replicates are modeled by the gamma distribution.
 - Within each biological replicate, you have tag counts for specific genes. The variance of these genes are modeled by a poisson distribution
 
This means that the tag counts of each gene is actually modeled by a mixture of the gamma (biological replicate) and poisson (tag count) distributions.

This effectively just means that the tag counts are modeled with a Poission distribution with the addition of extra variance defined by the biological replicates.
