# M. lignano codon usage

To establish codon usage, the Trinity stranded assembly on PE reads were used. 12,605 transcripts that had a complete ORF (start and stop) longer than 100 AA and also had blast hit with 1e-50 to proteomes were kept for generating codon usage table.

Codon usage table was generated using EMBOSS' [cusp](http://emboss.sourceforge.net/apps/cvs/emboss/apps/cusp.html).

I also took 14 complete CDS transcripts for M. lignano on pubmed and made a codon usage table as a comparison.

The plots below shows ruby, clover, EGFP, EGFP variant codon usage across the sequence. Each bar shows the percent codon usage for the amino acid.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from collections import defaultdict
tableFile = open('/home/share/projects/mlig_transcriptome/codonUsage/stranded.orf.nuc.longest.300.blast.cusp')
tableFile.next()
tableFile.next()
tableFile.next()
tableFile.next()
tableFile.next()
tableFile.next()
tableFile.next()
usage = {}
optimalUsage = {}

for line in tableFile:
    meta = line.strip().split()
    usage[meta[0]] = (float(meta[2]) * 100,meta[1])
    if not optimalUsage.has_key(meta[1]):
        optimalUsage[meta[1]] = (meta[0], float(meta[2]) * 100)
    else:
        if optimalUsage[meta[1]][1] < float(meta[2]) * 100:
            optimalUsage[meta[1]] = (meta[0], float(meta[2]) * 100)
    
tableFile.close()

tableFile = open('/home/share/projects/mlig_transcriptome/codonUsage/mlig.known.mRNA.cusp')
tableFile.next()
tableFile.next()
tableFile.next()
tableFile.next()
tableFile.next()
tableFile.next()
tableFile.next()
tableFile.next()
usageUniprot = {}

for line in tableFile:
    meta = line.strip().split()
    usageUniprot[meta[0]] = (float(meta[2]) * 100,meta[1])
tableFile.close()
```

```python
def codon(seq):
    c = []
    for i in range(0,len(seq),3):
        c.append(seq[i:i+3])
    return c

def optimalCodon(seq, usage, optTable):
    aa = [usage[x][1] for x in codon(seq)]
    return [optTable[x] for x in aa]

def usageStat(title, seq, table, opt):
    seqCodon = codon(seq)[1:-1]
    seqUsage = [table[x][0] for x in seqCodon]
    optSeq = ''.join([x[0] for x in optimalCodon(seq, table, opt)])
    optCodon = codon(optSeq)[1:-1]
    optUsage = [table[x][0] for x in optCodon]
    
    diffUsage = []
    for i in range(len(seqCodon)):
        if seqCodon[i] != optCodon[i]:
            diffUsage.append(optUsage[i] - seqUsage[i])
    
    print title
    print 'Number of aa:', len(seq) / 3
    print 'Mean usage fraction:', sum(seqUsage) / float(len(seqUsage))
    print 'Number of non-optimal codons:', len(diffUsage)
    print 'Mean usage difference from optimal:', sum(diffUsage) / float(len(diffUsage))
    print
    print 'Optimal codon usage sequence:'
    print optSeq
    print 'Mean optimal usage fraction:', sum(optUsage) / float(len(optUsage))
    fig, axes = plt.subplots(nrows = 2, ncols=1)
    pd.Series(seqUsage).plot(kind = 'bar', figsize = [50,10], ax = axes[0])
    pd.Series(optUsage).plot(kind = 'bar', ax = axes[1])
    axes[0].set_title('Usage fractions per amino acid')
    axes[1].set_title('Optimal usage fractions per amino acid')
    fig.tight_layout()
    
```
