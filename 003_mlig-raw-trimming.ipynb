# Trimming raw M. lignano data

The pair-end pilot libraries will be used for sequencing. All libraries except the H1 and H2 long insert libraries (1000 length fragments) were 350 length fragments. The reads are 100 bp. 

Trimmomatic will be used to trim the pilot reads. I will trim to a minimum length of 65 and minimum quality of 30.

The SE regeneration libraries are 50 bp libraries. These will be trimmed to a minimum length of 40 and minimum quality of 30.

```python
import gzip, os, sys
#import custom functions for displaying tables, bash commands
sys.path.append(os.path.abspath("/home/damian/"))
from dk_ipython import *
from IPython.display import HTML
```

```python
fs = [x for x in os.listdir('/home/share/ngsData/mlig/') if x.split('.')[-1] == 'gz' and x.find('pilot') != -1]

trimCmdPEA = 'java -jar /mlig_raw/software/Trimmomatic-0.32/trimmomatic-0.32.jar PE -threads 1 '
trimCmdB = ' LEADING:30 TRAILING:30 MINLEN:65 ILLUMINACLIP:/mlig_raw/software/Trimmomatic-0.32/adapters/adapters.fa:2:30:12:1:false TOPHRED33'
trimCmds = ''
for lib in list(set(['_'.join(f.split('_')[:-1]) for f in fs if f.split('.')[0].split('_')[-1][0] == 'R'])):
    a = lib + '_R1.fastq.gz'
    b = lib + '_R2.fastq.gz'
    phred = detectPhred('/home/share/ngsData/mlig/' + a, 5000)
    
    if phred == 'Sanger':
        phred = '-phred33 '
    else:
        phred = '-phred64 '
        
    a = '/mlig_raw/' + lib + '_R1.fastq.gz'
    b = '/mlig_raw/' + lib + '_R2.fastq.gz'
    #paired end
    trimCmds += trimCmdPEA + phred + a + ' ' + b + ' ' + \
    a + '.paired.trimmed.fastq.gz ' + a + '.unpaired.trimmed.fastq.gz ' + \
    b + '.paired.trimmed.fastq.gz ' + b + '.unpaired.trimmed.fastq.gz' + trimCmdB + '\n'
HTML(textarea('Trimmomatic bash commands for pilot reads',trimCmds))
```

```python
fs = [x for x in os.listdir('/home/share/ngsData/mlig/') if x.split('.')[-1] == 'gz' and x.find('pilot') == -1]

trimCmdSEA = 'java -jar /mlig_raw/software/Trimmomatic-0.32/trimmomatic-0.32.jar SE -threads 1 '
trimCmdB = ' LEADING:30 TRAILING:30 MINLEN:40 ILLUMINACLIP:/mlig_raw/software/Trimmomatic-0.32/adapters/adapters.fa:2:30:12:1:false TOPHRED33'
trimCmds = ''

for lib in [f for f in fs if f.split('.')[0].split('_')[-1][0] != 'R']:
    lib = lib.split('.')[0]
    phred = detectPhred('/home/share/ngsData/mlig/' + lib + ".fastq.gz", 5000)
    if phred == 'Sanger':
        phred = '-phred33 '
    else:
        phred = '-phred64 '
        
    trimCmds += trimCmdSEA + phred + '/mlig_raw/' + lib + '.fastq.gz ' + lib + '.trimmed.fastq.gz' + trimCmdB + '\n'
    
HTML(textarea('Trimmomatic bash commands for regeneration reads',trimCmds))
```

### Trim results

```python
trimmLib = {}
prev = ''
trimmLog = open('resource/trimPE.err')
for line in trimmLog:
    if line.find('Started with arguments') != -1:
        prev = line.strip().split('.fastq.gz')[0].split()[-1].split('/')[-1]
    if line.find('Surviving') != -1:
        meta = line.strip().split(':')
        numReads = int(meta[1].split()[0])
        keys = ['Reads'] + [meta[1].split()[-1]] + [' '.join(x.split()[2:]) for x in meta[2:-1]]
        vals = [(numReads,100)] + [(int(x.split()[:2][0]),float(x.split()[:2][1][1:-2])) for x in meta[2:]]
        kv = dict(zip(keys,vals))
        trimmLib[prev] = kv

trimmTable = ListTable()
for k,v in trimmLib.items():
    trimmTable.append([str(x) for x in [k,commas(v['Reads'][0]),v['Dropped'][1],v['Surviving'][1],v['Forward Only Surviving'][1],v['Reverse Only Surviving'][1]]])
    
trimmTable.sort(key = lambda x:x[0])
trimmTable.insert(0,['Library','Total fragments','Dropped(%)','Both Surviving(%)','Forward Only Surviving(%)', 'Reverse Only Surviving(%)'])
trimmTable
```
