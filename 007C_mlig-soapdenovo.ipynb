### Assembly metrics

````python
prefix = '/home/share/projects/mlig_transcriptome/soapdenovo/'

fs = [x for x in os.listdir(prefix) if x.split('.')[-1] == 'scafSeq']
      
from collections import defaultdict
from Bio import SeqIO

lengths = defaultdict(list)
for f in fs:
    inFile = open(prefix + f)
    k = 'k' + f.split('.')[1]
    
    for record in SeqIO.parse(inFile,'fasta'):
        l = len(str(record.seq))
        lengths[k].append(l)
    inFile.close()
````

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

lengthItems = lengths.items()
lengthItems.sort(key = lambda x : x[0])

def n50(ls):
    ls.sort(reverse = True)
    half = sum(ls) / 2
    total = 0
    for l in ls:
        total += l
        if total > half:
            return l
    
metricTable = ListTable()
metricTable.append(['k-value','transcripts #','mean length','median length','N50 length'])
for k, l in lengthItems:
    s = pd.Series(l)
    metricTable.append([k,commas(len(s)),s.mean(),s.median(),n50(l)])
metricTable
````

````python
numCol = 2
numRow = 4
fig, axes = plt.subplots(ncols = numCol, nrows = numRow, figsize=[20, 20])
i = 0
for k, l in lengthItems:
    r = i / numRow
    c = i % numRow
    l = pd.Series(l)
    l[l < 500].hist(ax = axes[c,r], bins=100)
    axes[c,r].set_title(k)
    axes[c,r].set_xlim(100, 500)
    axes[c,r].set_ylim(0,30000)
    i += 1
fig.tight_layout()
    
````

### Large number of N's in assemblies

````python
prefix = '/home/share/projects/mlig_transcriptome/soapdenovo/'

fs = [x for x in os.listdir(prefix) if x.split('.')[-1] == 'scafSeq']

Npercent = defaultdict(list)
for f in fs:
    inFile = open(prefix + f)
    k = 'k' + f.split('.')[1]
    
    for record in SeqIO.parse(inFile,'fasta'):
        numN = str(record.seq).lower().count('n')
        total = float(len(str(record.seq)))
        percentN = numN / total
        
        Npercent[k].append(percentN)
    inFile.close()
````

````python
NpercentItems = Npercent.items()
NpercentItems.sort(key = lambda x : x[0])

metricTable = ListTable()
metricTable.append(['k-value','transcripts #','mean N percent','# transcripts > 20% Ns','# transcripts > 25% Ns','# transcripts > 30% Ns'])
for k, l in NpercentItems:
    s = pd.Series(l)
    
    metricTable.append([k,commas(len(s)),s.mean() * 100,commas(len(s[s > 0.2])),commas(len(s[s > 0.25])),commas(len(s[s > 0.30]))])
metricTable
````

````python
numCol = 2
numRow = 4
fig, axes = plt.subplots(ncols = numCol, nrows = numRow, figsize=[20, 20])
i = 0
for k, l in NpercentItems:
    r = i / numRow
    c = i % numRow
    l = pd.Series(l)
    l[l > 0].hist(ax = axes[c,r], bins=100)
    axes[c,r].set_title(k)
    i += 1
fig.tight_layout()
````
