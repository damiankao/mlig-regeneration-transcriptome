To check whether jaccard clipping made a difference. A fusion score was calculated for each transcript. Fusion score is the ratio of longest blast alignment to the union of all blast alignments. A low fusion score potentially can mean a chimeric transcript. 

The following scripts were used to generate fusion scores:

 1. We first consolidated all 4 assemblies into one file and renamed each transcript for easier blast against complete proteomes.
 2. We blasted this file against proteomes with the following command:
         blastx -query combined.fasta -db cele.csin.dmel.emul.hsap.sman.proteom -evalue 1e-5 -outfmt '6 qseqid sseqid qlen slen pident length qstart qend sstart send evalue' -num_threads 30 > jaccard.blast.out
 3. We then format the blast output into two different bed files displaying hits and HSP. We want to group union of alignments by hits. And we want to find the longest alignment by union of HSPs for a hit.
 4. Run bedtools merge on hit and HSP bed files.
 5. Use the provided script to calculate the fusion scores.

```python
HTML(textarea('Script for renaming and consolidating the jaccard/noJaccard contig files','''import sys,os

strandedFile = open(sys.argv[1],'r')
strandedJaccardFile = open(sys.argv[2],'r')
unstrandedFile = open(sys.argv[3],'r')
unstrandedJaccardFile = open(sys.argv[4],'r')

outFile = open('combined.fasta','w')

def writeFile(inStream, outStream, prefix):
    count = 0
    for line in inStream:
        if line[0] == '>':
            outStream.write('>' + prefix + '.' + str(count) + '\\t' + line[1:])
            count += 1
        else:
            outStream.write(line)
        
writeFile(strandedFile, outFile, 'stranded')
writeFile(strandedJaccardFile, outFile, 'strandedJaccard')
writeFile(unstrandedFile, outFile, 'unstranded')
writeFile(unstrandedJaccardFile, outFile, 'unstrandedJaccard')

outFile.close()

'''))
```

```python
#format the blast file into a sorted .bed file for bedtools merge
HTML(textarea('Create query-subject .bed file for HSP unions','''import sys
from collections import defaultdict

jaccardBlast = open(sys.argv[1])

transcript = defaultdict(list)
for line in jaccardBlast:
    data = line.strip().split()
    transcript[data[0] + '.' + data[1]].append((data[-3],data[-2],data[1],data[-1]))
    
tids = transcript.keys()
#sort by "chromosome"
tids.sort()

for tid in tids:
    feats = transcript[tid]
    #sort by start coodinate
    feats.sort(key = lambda x : int(x[0]))
    for feat in feats:
        print tid + '\\t' + '\\t'.join(feat)'''))
```

```python
#format the blast file into a sorted .bed file for bedtools merge
HTML(textarea('Create query .bed file for hit unions','''import sys
from collections import defaultdict

jaccardBlast = open(sys.argv[1])

transcript = defaultdict(list)
for line in jaccardBlast:
    data = line.strip().split()
    transcript[data[0]].append((data[-3],data[-2],data[1],data[-1]))
    
tids = transcript.keys()
#sort by "chromosome"
tids.sort()

for tid in tids:
    feats = transcript[tid]
    #sort by start coodinate
    feats.sort(key = lambda x : int(x[0]))
    for feat in feats:
        print tid + '\\t' + '\\t'.join(feat)'''))
```

```python
prefix = '/home/share/projects/mlig_transcriptome/assemblyAnalaysis/trinity/checkJaccard/'
hitFile = open(prefix + 'jaccard.blast.hit.merged.bed')
hspFile = open(prefix + 'jaccard.blast.hsp.merged.bed')

unions = defaultdict(int)
lengths = defaultdict(lambda : defaultdict(int))

for line in hitFile:
    meta = line.strip().split()
    unions[meta[0]] += int(meta[2]) - int(meta[1]) + 1
    
for line in hspFile:
    meta = line.strip().split()
    q = '.'.join(meta[0].split('.')[:2])
    s = meta[0].split('.')[2]
    lengths[q][s] += int(meta[2]) - int(meta[1]) + 1    

hitFile.close()
hspFile.close()

st_jaccardScore = []
st_noJaccardScore = []
st_jaccardIndex = []
st_noJaccardIndex = []

un_jaccardScore = []
un_noJaccardScore = []
un_jaccardIndex = []
un_noJaccardIndex = []

for q, ls in lengths.items():
    fscore = max(ls.values()) / float(unions[q])
    qid = q.split('.')[0]
    if qid == 'strandedJaccard':
        st_jaccardScore.append(fscore)
        st_jaccardIndex.append(q)
    elif qid == 'stranded':
        st_noJaccardScore.append(fscore)
        st_noJaccardIndex.append(q)
    elif qid == 'unstrandedJaccard':
        un_jaccardScore.append(fscore)
        un_jaccardIndex.append(q)
    elif qid == 'unstranded':
        un_noJaccardScore.append(fscore)
        un_noJaccardIndex.append(q)
```

```python
st_jaccardScore = pd.Series(st_jaccardScore, index = st_jaccardIndex)
st_noJaccardScore = pd.Series(st_noJaccardScore, index = st_noJaccardIndex)
un_jaccardScore = pd.Series(un_jaccardScore, index = un_jaccardIndex)
un_noJaccardScore = pd.Series(un_noJaccardScore, index = un_noJaccardIndex)
```

```python
print 'Distribution of fusion scores in the 4 assembly variants'
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = [15,5])
st_noJaccardScore.hist(bins = 100, ax = axes[0,0])
st_jaccardScore.hist(bins = 100, ax = axes[0,1])
un_noJaccardScore.hist(bins = 100, ax = axes[1,0])
un_jaccardScore.hist(bins = 100, ax = axes[1,1])

axes[0,0].set_title('stranded no jaccard clipping fusion score distribution')
axes[0,1].set_title('stranded jaccard clipping fusion score distribution')
axes[1,0].set_title('unstranded no jaccard clipping fusion score distribution')
axes[1,1].set_title('unstranded jaccard clipping fusion score distribution')

fig.tight_layout()
```

### SOAPDeNovo-trans assemblies

8 assemblies were performed with k-value from 25 - 60 at increments of 5. All PE and SE reads were used in these assemblies. Since SOAPDeNovo-trans does not have a strandedness option, all reads were used as unstranded.

```python
from Bio import SeqIO

prefix = '/home/share/projects/mlig_transcriptome/soapdenovo/'

fs = [x for x in os.listdir(prefix) if x.split('.')[-1] == 'scafSeq']

soap_assembly = defaultdict(list)

for f in fs:
    k = f.split('.')[1]
    for record in SeqIO.parse(prefix + f,'fasta'):
        seq = str(record.seq)
        desc = record.description.split()
        l = len(seq)
        t = desc[-1]
        n = seq.count('N')
        np = n / float(l)
        soap_assembly[k].append((l,n,np,t))
```

```python
soap_dfs = {}
for k, d in soap_assembly.items():
    soap_dfs[k] = pd.DataFrame(d)
    soap_dfs[k].columns = ['length','ambig','ambig p','type']
```

```python
def metricTable(dfs, keys):
    out = ListTable()
    out.append(['k-value','# of transcripts','mean length','median length','N50 length'])
    for i, k in enumerate(keys):
        df = dfs[k]
        half = df['length'].sum() / 2
        count = 0
        n50 = 0
        sorted_df = df['length'].copy()
        sorted_df.sort(ascending = False)
        
        for index, value in sorted_df.iteritems():
            count += value
            if count > half:
                n50 = value
                break
                
        out.append([k,commas(len(df)),df['length'].mean(),df['length'].median(), commas(int(n50))])
    return out

def lengthDistro(dfs, keys, r, c, w = 15, h = 15):
    fig, axes = plt.subplots(ncols = c, nrows = r, figsize=[w,h])

    for i, k in enumerate(keys):
        df = dfs[k]
        row = i % r
        col = i / r
        df[df['length'] < 1000]['length'].hist(bins=150, ax = axes[row, col])
        axes[row,col].set_title(k)

    fig.tight_layout()
    
def ambigTable(dfs, keys):
    out = ListTable()
    out.append(['k-value','# of transcripts with %N > 0','% of total transcripts with %N > 0','mean %N','median %N','mean length', 'median length'])
    for i, k in enumerate(keys):
        df = dfs[k]
        df = df[df['ambig p'] > 0]
        out.append([k,commas(len(df)),len(df) / float(len(dfs[k])) * 100,df['ambig p'].mean(),df['ambig p'].median(), df['length'].mean(), df['length'].median()])
    return out

def lengthNHexbin(dfs, keys, r, c):
    fig, axes = plt.subplots(ncols = c, nrows = r, figsize=[15,20])

    for i, k in enumerate(keys):
        df = soap_dfs[k]
        row = i % r
        col = i / r
        df[df['ambig p'] > 0][df['length'] < 5000].plot(kind='hexbin',x = 'length',y = 'ambig p', ax = axes[row, col])
        axes[row,col].set_title(k)

    fig.tight_layout()
```

```python
metricTable(soap_dfs, ['25','30','35','40','45','50','55','60'])
```

```python
lengthDistro(soap_dfs, ['25','30','35','40','45','50','55','60'], 4, 2)
```

```python
ambigTable(soap_dfs, ['25','30','35','40','45','50','55','60'])
```

```python
lengthNHexbin(soap_dfs, ['25','30','35','40','45','50','55','60'], 4, 2)
```
